```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

#2
X = np.array([[1, 0], [0, 1], [0, -1], [-1, 0], [0, 2], [0, -2], [-2, 0]])
Y = np.array([-1, -1, -1, 1, 1, 1, 1])
row, col = X.shape
z = np.zeros(X.shape)
z[:, 0] = 2*X[:, 1]**2-4*X[:, 0]+2
z[:, 1] = X[:, 0]**2-2*X[:, 1]-1
pos1 = Y == -1; pos2 = Y == 1
plt.plot(z[pos1, 0], z[pos1, 1], 'x', z[pos2, 0], z[pos2, 1], 'o')
plt.show()
#l1x = [4, 4]; l1y = [-7.5, 1.5]
#l2x = [5, 5]; l2y = [-7.5, 1.5]
#plt.plot(l1x, l1y, '--',l2x, l2y, '--')

#3
clf = SVC(C=1e10, kernel='poly', degree=2, gamma=2, coef0=1, shrinking=False)
clf.fit(X, Y)
print(clf.dual_coef_, clf.intercept_, clf.support_)
```

```python
## Drawing the beta distribution
import sys
import scipy as sp
from scipy.special import gamma
import pylab


a = 2
b = 10


def pdf(a, b, x):
    Z = gamma(a + b) / gamma(a) / gamma(b)
    return x**(a - 1) * (1 - x)**(b - 1) / Z


x = sp.arange(0.0, 1.01, 0.01)
y = pdf(a, b, x)
ax = pylab.subplot(111)
ax.plot(x, y)

pylab.show()
```
